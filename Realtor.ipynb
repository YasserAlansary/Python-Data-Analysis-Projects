```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

df = pd.read_excel('F:\\New folder\\Realtor.xlsx')
df
```

```python
df = df.fillna(df.median())
df
```

```python
df['price per sqft'] = round(df['price']/df['house_size'],2)
df.head(100)
```

```python
df.describe()
```

```python
#Average price per state
state = df.groupby('state')
states = state.agg({'price':'mean'})
States = states.sort_values('price', ascending = True)
States.plot(kind = 'bar', color='k' ,alpha = 0.8)
plt.title('Average price per state')
plt.show
```

```python
#Number of houses for each state 
state = df.groupby('state')['full_address'].count().reset_index(name='number')
State = state.sort_values(by = 'number', ascending = True)
States = State.set_index('state')
States.plot(kind = 'bar',color='b',alpha = 0.8)
plt.title('Number of houses for each state ')
plt.show()
```

```python
#The top 30 cities in terms of the number of homes sold or offered for sale
city = df.groupby('city')['full_address'].count().reset_index(name='number')
City = city.sort_values(by = 'number', ascending = True)
Cities = City.set_index('city')
Cities1 = Cities[-30:]
Cities1.plot(kind = 'bar', color='b', alpha = 0.8)
plt.title('Top 30 cities in terms of number of homes')
plt.show()
```

```python
#Most expensive cities over a country
city = df.groupby('city')['price'].mean().reset_index(name='price')
cities = city.sort_values(by = 'price', ascending = True)
Cities = cities.set_index('city')
top = Cities[-20:]
top.plot(kind = 'barh', color='k' ,alpha = 0.8)
plt.title('Most expensive cities')
plt.show
```

```python
#the cheapest cities over a country 
bottom = Cities[:20]
bottom.plot(kind = 'barh', color='k' ,alpha = 0.8)
plt.title('the cheapest cities')
plt.show
```

```python
#Best cities if you looking for big house 
data = df.groupby('city')['house_size'].mean().reset_index(name='size')
cities = data[data['size'] >= 2100]
Cities1 = cities.sort_values(by = 'size', ascending = True)
Cities2 = Cities1.set_index('city')
best = Cities2[:20]
best.plot(kind = 'barh', color='b' ,alpha = 0.8)
plt.title('Best cities for big house')
plt.show
```
