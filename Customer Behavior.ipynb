```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
customers = pd.read_excel('Shopping_Behavior.xlsx')
customers
```

```python
customers.describe()
```

```python
#Total spend money for each age group
ages= customers['Age'] / 10

age_range= []

for num in ages: 
    if num >= 1 and num < 2:
            age_range.append('teenagers')
    elif num >= 2 and num < 3.4:
            age_range.append('adult')
    elif num >= 3.4 and num < 4.8:
            age_range.append('senior')
    else:
            age_range.append('old')

customers['Age Range']= age_range

age= customers.groupby('Age Range').agg({'Purchase Amount':'sum'}).reset_index()
Age= age.sort_values('Purchase Amount',ascending= False)

Age
```

```python
age= customers.groupby('Age Range').agg({'Review Rating':'mean'}).reset_index().rename(columns={'Review Rating':'Rate'})
Age= age.sort_values('Rate',ascending= False).round(2)

Age
```

```python
#Total spend money for each gender
amount= customers.groupby('Gender').agg({'Purchase Amount':'sum'}).reset_index()
Amount= amount.sort_values('Purchase Amount',ascending= False)

Amount
```

```python
#Total Sales for each item 
item= customers.groupby(['Item Purchased']).agg({'Purchase Amount':'sum'
                                                    }).reset_index().rename(columns={'Purchase Amount':'Sales'})

Item= item.sort_values('Sales',ascending= False)
Item
```

```python
#Total Sales for each category 
cate= customers.groupby('Category').agg({'Purchase Amount':'sum'}).reset_index().rename(columns={'Purchase Amount':'Sales'})

Cate= cate.sort_values('Sales',ascending= False)

Cate
```

```python
#Common size for each product
item= customers.groupby(['Item Purchased','Size']).agg({'Purchase Amount':'sum'
                                                    }).reset_index().rename(columns={'Purchase Amount':'Sales'})

Item= item.sort_values('Sales',ascending= False)
Item
```

```python
#Average rating for each product
item= customers.groupby(['Item Purchased']).agg({'Review Rating':'mean'
                                                    }).reset_index().rename(columns={'Review Rating':'Rate'})

Item= item.sort_values('Rate',ascending= False).round(1)
Item
```

```python
#Average rating for each category 
cate= customers.groupby('Category').agg({'Review Rating':'mean'}).reset_index().rename(columns={'Review Rating':'Rate'})

Cate= cate.sort_values('Rate',ascending= False).round(2)

Cate
```

```python
#total sales for each season
season= customers.groupby('Season').agg({'Purchase Amount':'sum'}).reset_index().rename(columns={
    'Purchase Amount':'Sales'})

Season= season.sort_values('Sales',ascending= False)
Season
```

```python
#Common products in every season
season= customers.groupby(['Season','Item Purchased']).agg({'Purchase Amount':'sum'}).reset_index().rename(columns={
    'Purchase Amount':'Sales'})

Season= season.sort_values(['Season','Sales'],ascending= False)
Season
```

```python
#Common colors in each season
color= customers.groupby(['Season','Color']).agg({'Item Purchased':'count'}).reset_index().rename(columns={
    'Item Purchased':'Product'})

Color= color.sort_values(['Season','Product'],ascending= False)
Color
```

```python
#Total Sales and customers in each city
city= customers.groupby(['Location']).agg({'Customer ID':'count','Purchase Amount':'sum'}).reset_index().rename(
    columns={'Purchase Amount':'Sales'})

City= city.sort_values('Sales',ascending= False)
City
```

```python
#Satisfaction level in each city
city= customers.groupby(['Location']).agg({'Review Rating':'mean'}).reset_index().rename(
    columns={'Review Rating':'Rate'})

City= city.sort_values('Rate',ascending= False).round(2)
City
```

```python
#Total sales based on subscription status
sub= customers.groupby('Subscription Status').agg({'Customer ID':'count','Purchase Amount':'sum'}).reset_index().rename(
    columns={'Customer ID':'Customers'})

Sub= sub.sort_values('Purchase Amount',ascending= False)
Sub
```

```python
#Subscription effect 
sub= customers.groupby(['Category','Subscription Status']).agg({'Customer ID':'count',
    'Purchase Amount':'sum'}).reset_index().rename(columns={'Customer ID':'Customers'})

Sub= sub.sort_values(['Subscription Status','Purchase Amount'],ascending= False)
Sub
```

```python
#Promo code effect
sub= customers.groupby('Promo Code Used').agg({'Customer ID':'count','Purchase Amount':'sum'}).reset_index().rename(
    columns={'Customer ID':'Customers'})

Sub= sub.sort_values('Purchase Amount',ascending= False)
Sub
```

```python
#Popular shipping companies  
sub= customers.groupby('Shipping Type').agg({'Customer ID':'count'}).reset_index().rename(
    columns={'Customer ID':'Customers'})

Sub= sub.sort_values('Customers',ascending= False)
Sub
```

```python
#Favorite payment method and total transformed money 
sub= customers.groupby('Payment Method').agg({'Customer ID':'count','Purchase Amount':'sum'}).reset_index().rename(
    columns={'Customer ID':'Customers','Purchase Amount':'Transferred funds'})

Sub= sub.sort_values('Customers',ascending= False)
Sub
```

```python
#Frequency of Purchases
sub= customers.groupby(['Frequency of Purchases']).agg({'Purchase Amount':'sum'}).reset_index()

Sub= sub.sort_values(['Purchase Amount'],ascending= False)
Sub
```

```python
#Discount effect on frequency of purchases
sub= customers.groupby(['Frequency of Purchases','Discount Applied']).agg({'Purchase Amount':'sum'}).reset_index()

Sub= sub.sort_values(['Discount Applied','Purchase Amount'],ascending= False)
Sub
```

```python
#Frequency of purchases for each gender 
sub= customers.groupby(['Gender','Frequency of Purchases']).agg({'Customer ID':'count'}).reset_index().rename(
    columns={'Customer ID':'Customers'})

Sub= sub.sort_values(['Gender','Customers'],ascending= False)
Sub
```
