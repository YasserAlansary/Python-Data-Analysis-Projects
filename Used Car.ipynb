```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

df= pd.read_excel('F:\\New folder\\Used_Cars.xlsx')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
#Average price for each car
name = df.groupby('name')['selling_price'].mean().reset_index(name = 'price')
car = name.sort_values(by = 'price')
Car = car.set_index('name')
Car.plot(kind ='bar', color='k',alpha=0.8 )
plt.title('Average price for each brand')
plt.show()
```

```python
#Max power average for each brand
brand = df.groupby('name')['max_power'].mean().reset_index(name = 'Max_power')
car = brand.sort_values(by = 'Max_power')
Car = car.set_index('name')
Car.plot(kind ='bar', color='k',alpha=0.8 )
plt.title('Average max powerfor each brand')
plt.show()
```

```python
#Engine power average for each brand
brand = df.groupby('name')['engine'].mean().reset_index(name = 'engine')
car = brand.sort_values(by = 'engine')
Car = car.set_index('name')
Car.plot(kind ='bar', color='k',alpha=0.8 )
plt.title('Average engine power for each brand')
plt.show()
```

```python
#Average distance for each brand over years
brand = df.groupby('name')['km_driven'].mean().reset_index(name = 'distance')
car = brand.sort_values(by = 'distance')
Car = car.set_index('name')
Car.plot(kind ='bar', color='k',alpha=0.8 )
plt.title('Average distance for each brand')
plt.show()
```

```python
#Average price for each car per year
year = df.groupby('year')['selling_price'].mean().reset_index(name = 'price')
data = year.sort_values(by = 'year')
Year = data.set_index('year')
Year.plot(kind ='line', color='k',alpha=0.8, marker = 'o')
plt.title('Average price per year')
plt.show()
```

```python
region = df.groupby('Region')['name'].count().reset_index(name = 'numbers')
data = region.sort_values(by = 'numbers')
regions = data['Region'].tolist()
Regions = data['numbers'].tolist()
plt.pie(Regions, labels = regions, autopct='%1.1f%%',
       pctdistance=1.2, labeldistance=.6)
plt.title('Average distance for each brand')
plt.show()
```

```python
#The states with the highest number of cars sold
place = df.groupby('State')['name'].count().reset_index(name = 'numbers')
data = place.sort_values(by = 'numbers')
state = data.set_index('State')
State = state[-20:]
State.plot(kind ='barh', color='k',alpha=0.8 )
plt.title('The states with the highest number of cars sold')
plt.show()
```

```python
#The states with the highest number of modern cars 
new_df = df[df.transmission == 'Automatic']
place = new_df.groupby('State')['transmission'].count().reset_index(name = 'numbers')
data = place.sort_values(by = 'numbers')
state = data.set_index('State')
State = state[-20:]
State.plot(kind ='barh', color='k',alpha=0.8 )
plt.title('The states with the highest number of modern cars')
plt.show()
```

```python
#The total number of cars for each brand in each state
data = df.groupby(['State','name'])['name'].count().reset_index(name = 'numbers')
state = data[data.numbers > 10]
State = state.sort_values('numbers').reset_index()
State.drop('index', axis = 1, inplace= True)
State
```

```python
#Total numbers for each fuel type
fuel_type = df.groupby('fuel')['name'].count().reset_index(name = 'numbers')
data = fuel_type.sort_values(by = 'numbers')
Fuel_type = data.set_index('fuel')
Fuel_type.plot(kind = 'bar', color='k', alpha= 0.7)
plt.title('The states with the highest number of cars sold')
plt.show()
```

```python
#The average price of a car according to the type of fuel it consumes and the model year
fuel_type = df.groupby(['fuel','year']) 
price = fuel_type.agg({'selling_price':'mean'})
price
```

```python
#Number of cars in each transmission category 
data = df.groupby('transmission')['name'].count().reset_index(name ='numbers')
transmission = data.sort_values(by = 'numbers')
Transmission= transmission.set_index('transmission')
Transmission.plot(kind = 'barh',color= 'k', alpha= 0.7)
plt.show()
```

```python
#Average car price by transmission type and model year
transmission = df.groupby(['transmission','year']) 
price = transmission.agg({'selling_price':'mean'})
price
```

```python
#number of cars in each owner type and average of driven distance and price
data = df.groupby('owner')
owner = data.agg({'name':'count','km_driven':'mean','selling_price':'mean'}).reset_index().rename(columns={'name':'Brands',
'km_driven':'Total Distance','selling_price':'Price'})
owner
```

```python
#Average of engine power, max power and price for each brand based on model year 
names = df.groupby(['name','year'])
data = names.agg({'engine':'mean','max_power':'mean','selling_price':'mean'})
data
```

```python
#verage price for each seller type
seller_type= df.groupby('seller_type')['selling_price'].mean().reset_index(name= 'Price')
Seller_type = seller_type.sort_values('Price')
Price = Seller_type.set_index('seller_type')
Price.plot(kind='barh', color= 'k', alpha= 0.7)
plt.title('Average price for each seller type')
plt.show()
```
